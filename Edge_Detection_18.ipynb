```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img=cv2.imread('DATA/sammy_face.jpg')
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
med=np.median(img)
print(med)
```

```python
lower = int(max(0,0.7*med))
upper = int(min(255,1.3*med))
```

```python
dst=cv2.Canny(img,threshold1=lower,threshold2=upper)
```

```python
plt.imshow(dst)
```

```python
img_blur=cv2.blur(img,ksize=(5,5))
```

```python
dst=cv2.Canny(img_blur,threshold1=lower,threshold2=upper)
```

```python
plt.imshow(dst)
```

```python
dst=cv2.Canny(img_blur,threshold1=lower,threshold2=upper+50)
plt.imshow(dst)
```
